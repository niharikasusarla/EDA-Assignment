```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
file_path = '/content/drive/MyDrive/covid_19_india.csv'
covid_data = pd.read_csv(file_path)
```

```python
print("First 5 rows of the dataset:")
print(covid_data.head())
```

```python
print("\nMissing values in each column:")
print(covid_data.isnull().sum())
```

```python
print("\nDataset Summary:")
print(covid_data.info())
```

```python
print("\nBasic Statistics for Numerical Columns:")
print(covid_data.describe())
```

```python

# 1. Total confirmed cases by state
statewise_cases = covid_data.groupby('State/UnionTerritory')['Confirmed'].max().sort_values(ascending=False).head(10)
statewise_cases.plot(kind='bar', color='blue', title='Top 10 States with Highest Confirmed Cases')
plt.xlabel('State')
plt.ylabel('Confirmed Cases')
plt.show()
```

```python
# 2. Distribution of Deaths
sns.histplot(covid_data['Deaths'], kde=True, color='blue')
plt.title('Distribution of Deaths')
plt.xlabel('Deaths')
plt.show()
```

```python

# 3. Trend of confirmed cases over time
covid_data['Date'] = pd.to_datetime(covid_data['Date'])
datewise_cases = covid_data.groupby('Date')['Confirmed'].sum()
datewise_cases.plot(color='green', title='Trend of Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.show()
```
